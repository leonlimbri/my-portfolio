## Monty Hall
Monty hall problem is a probability puzzle where given a choice of 3 doors to a contestant of a game show, where behind 2 doors are a **goat** and 1 of the door has a **car** in it. The story begins with the contestant given a choice to choose. Then, the host of the show who knows what's behind the doors, opens another door, then give you another choice to move to another door. The problem statement is that _Is it to your advantage to switch your choice?_

This jupyer notebook will simulate this situation `n` times and conclude this problem.

#### Packages to be imported

```python
import random
import matplotlib.pyplot as plt
import matplotlib.animation
from matplotlib.animation import FuncAnimation
from IPython import display
from IPython.display import HTML
```

#### Functions
Generate a function to create an array of spiralling numbers of size n (without denoting whether the number is prime or not).

```python
def play_monty_hall(switch = False) -> bool:
    '''
        Play the Monty Hall game show once.
        
        Parameters
        ----------
        switch: bool
            A boolean that affects the choice of the contestant
            when given the option to switch.
            
        Returns
        -------
        bool
            A boolean that shows whether the contestant
            win or lose the game.
            
    '''
    
    # Simulate the game by putting the car in a random place
    doors                       = ['goat', 'goat', 'goat']
    doors[random.randint(0, 2)] = 'car'
    
    init_position = random.randint(0, 2)
    pos_          = init_position
    
    # Opening the other door
    if doors[(init_position - 1) % 3]  == 'car':
        doors[(init_position + 1) % 3] = 'open'
    elif doors[(init_position + 1) % 3]  == 'car':
        doors[(init_position - 1) % 3] = 'open'
    else:
        doors[(init_position + random.randint(1, 2)) % 3] = 'open'
    
    # Switch position
    if switch:
        if doors[(pos_ - 1) % 3] == 'open':
            pos_ = (pos_ + 1) % 3
        elif doors[(pos_ + 1) % 3] == 'open':
            pos_ = (pos_ - 1) % 3
        
    return doors[pos_] == 'car'
```

```python
def animate(i):
    # Create the animation data.
    line_switch.set_data(range(1, i + 1), winning_switch_plot[:i])
    line_noswitch.set_data(range(1, i + 1), winning_noswitch_plot[:i])
```

#### Run the simulation

```python
n = 150
```

```python
winning_noswitch = [play_monty_hall(False) for i in range(n)]
winning_switch   = [play_monty_hall(True)  for i in range(n)]
```

```python
winning_noswitch_plot = [sum(winning_noswitch[:i])/(i + 1) * 100 for i in range(n)]
winning_switch_plot   = [sum(winning_switch[:i])/(i + 1) * 100 for i in range(n)]
```

```python
f, ax = plt.subplots()
ax.axis([1, n, 0, 100])
ax.set_xlabel('# Simulated Trial')
ax.set_ylabel('Winning Probability (%)')
ax.set_title('Winning Probability as number of simulation increases')
line_switch,   = ax.plot([], [])
line_noswitch, = ax.plot([], [])

ani = matplotlib.animation.FuncAnimation(f, animate, frames = n)
HTML(ani.to_jshtml())
```
